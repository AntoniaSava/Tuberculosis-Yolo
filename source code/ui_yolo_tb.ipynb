```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
from IPython.display import display
import ipywidgets as widgets
import os

MODEL_PATH = 'tb_yolo_seg/exp2/weights/best.pt'
model = YOLO(MODEL_PATH)

upload = widgets.FileUpload(accept='.jpg,.png', multiple=False)
run_button = widgets.Button(description='Segmentare')
output = widgets.Output()


def on_run_button_clicked(b):
    output.clear_output()
    if not upload.value:
        with output:
            print(" Te rog încarcă o imagine mai întâi.")
        return


    uploaded_file = upload.value[0]
    file_name = uploaded_file['name']
    with open(file_name, 'wb') as f:
        f.write(uploaded_file['content'])

   
    results = model(file_name)

   
    segmented_image = results[0].plot()

    output_filename = 'output_segmented.png'
    plt.imsave(output_filename, segmented_image[..., ::-1])  

   
    with output:
        print(f" Segmentare salvată ca: {output_filename}")
        plt.figure(figsize=(8, 8))
        plt.imshow(segmented_image[..., ::-1])
        plt.axis('off')
        plt.title("Segmentare TB cu YOLOv8")
        plt.show()


run_button.on_click(on_run_button_clicked)


display(widgets.VBox([upload, run_button, output]))
```
